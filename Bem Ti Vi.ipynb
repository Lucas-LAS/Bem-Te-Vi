```python
import pandas as pd  
#pd.set_option('float_format', '{:.0f}'.format)
pd.options.display.float_format = 'R$ {:,.2f}'.format
```

```python
df_Base_loja = pd.read_excel('Arquivo2BaseCaseDotz.xlsx',sheet_name ='Base_loja')
df_Base_loja_categoria = pd.read_excel('Arquivo2BaseCaseDotz.xlsx',sheet_name ='Base_loja_categoria')
```

```python
df_Base_loja.qtd_clientes=df_Base_loja.qtd_clientes.astype(int)
df_Base_loja.qtd_Tickets = df_Base_loja.qtd_Tickets.astype(int)
df_Base_loja.qtd_vendida = df_Base_loja.qtd_vendida.astype(int)
df_Base_loja.head(5)
```

```python
df_Base_loja_categoria.qtd_vendida = df_Base_loja_categoria.qtd_vendida.astype(int)
df_Base_loja_categoria.qtd_clientes = df_Base_loja_categoria.qtd_clientes.astype(int)
df_Base_loja_categoria.qtd_Tickets = df_Base_loja_categoria.qtd_Tickets.astype(int)
df_Base_loja_categoria.head()
```

```python
lista_mes={1:'Janeiro',2:'Fevereiro',3:'Março',4:'Abril',5:'Maio',6:'Junho',7:'Julho',8:'Agosto',9:'Setembro',10:'Outubro',11:'Novembro',12:'Dezembro'}
```

```python
df_Base_loja['Ano']=df_Base_loja['mês'].dt.year
df_Base_loja['Mês'] = df_Base_loja['mês'].dt.month
df_Base_loja.rename(columns={'mês':'Data_Completa'},inplace=True)
df_Base_loja['Nome Mês']=df_Base_loja['Mês'].map(lista_mes)
df_Base_loja.head()
```

```python
df_Base_loja_categoria['Ano']=df_Base_loja_categoria['mês'].dt.year
df_Base_loja_categoria['Mês'] = df_Base_loja_categoria['mês'].dt.month
df_Base_loja_categoria.rename(columns={'mês':'Data_Completa'},inplace=True)
df_Base_loja_categoria['Nome Mês']=df_Base_loja_categoria['Mês'].map(lista_mes)
df_Base_loja_categoria.head()
```

```python
df_Total_vendas=df_Base_loja.groupby(['loja','chavefilial'])[['valor_vendido','qtd_vendida','qtd_Tickets','qtd_clientes']]\
.agg('sum').sort_values(by='valor_vendido',ascending=False)

df_Total_vendas
```

```python
df_Total_vendas['Porcentagem_Total_Vendido']=(df_Total_vendas.valor_vendido/df_Total_vendas.valor_vendido.sum()) * 100
df_Total_vendas.Porcentagem_Total_Vendido = df_Total_vendas.Porcentagem_Total_Vendido.map('{:.2f}%'.format)

df_Total_vendas['Porcentagem_Quantidade_Vendido']=(df_Total_vendas.qtd_vendida/df_Total_vendas.qtd_vendida.sum()) * 100
df_Total_vendas.Porcentagem_Quantidade_Vendido = df_Total_vendas.Porcentagem_Quantidade_Vendido.map('{:.2f}%'.format)

df_Total_vendas['Porcentagem_Total_Tickets']=(df_Total_vendas.qtd_Tickets/df_Total_vendas.qtd_Tickets.sum()) * 100
df_Total_vendas.Porcentagem_Total_Tickets = df_Total_vendas.Porcentagem_Total_Tickets.map('{:.2f}%'.format)

df_Total_vendas['Porcentagem_Total_Clientes']=(df_Total_vendas.qtd_clientes/df_Total_vendas.qtd_clientes.sum()) * 100
df_Total_vendas.Porcentagem_Total_Clientes = df_Total_vendas.Porcentagem_Total_Clientes.map('{:.2f}%'.format)
df_Total_vendas.Porcentagem_Total_Clientes
```
